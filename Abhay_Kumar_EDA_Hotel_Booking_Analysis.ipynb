Exploratory Data Analysis (EDA) is a vital process that involves conducting preliminary investigations on data to uncover patterns, detect anomalies, test hypotheses, and validate assumptions. It employs summary statistics and graphical representations to gain valuable insights. The key steps involved in EDA are as follows:


*   **Data Collection:** Gathering the required data from reliable sources, ensuring its accuracy and completeness.

*   **Data Cleaning:** Removing or correcting any errors, missing values, or outliers in the dataset to ensure data quality.

*   **Data Visualization:** Creating visual representations such as scatter plots, histograms, and box plots to explore the distribution and relationships between variables.

*   **Descriptive Statistics:** Calculating summary statistics like mean, median, standard deviation, and quartiles to understand the central tendency and variability of the data.

*  **Feature Engineering:** Transforming or creating new variables based on existing ones to extract more meaningful insights.

*  **Correlation Analysis:** Examining the relationships between variables using correlation matrices or heatmaps to identify potential dependencies.

*  **Hypothesis Testing:** Conducting statistical tests to validate or reject hypotheses about the data.

*  **Data Interpretation:** Drawing conclusions and making inferences based on the observed patterns and statistical findings.

By performing EDA, researchers and analysts gain a deeper understanding of the dataset, uncover hidden trends, and make informed decisions based on the insights derived. It serves as a crucial foundation for further analysis and modeling.

````python
# Using Inter Quartile range in skew symmetric curve for removing outliers

# Outlier columns
outliers_columns=['lead_time','adr']

# Copy dataset as new dataset
new_df=df.copy()

# Capping dataset
for i in outliers_columns:
    #Findng IQR
    Q1=new_df[i].quantile(0.25)
    Q3=new_df[i].quantile(0.75)
    IQR=Q3-Q1

    # Defining lower and upper limit
    lower_limit =new_df[i].quantile(0.25)-1.5*IQR
    upper_limit =new_df[i].quantile(0.75)+1.5*IQR

    # Applying lower and upper limit to each variables
    new_df.loc[(new_df[i] > upper_limit),i] = upper_limit
    new_df.loc[(new_df[i] < lower_limit),i] = lower_limit
````

````python
# Obtaing target variable
excluded_variables=[var for var in new_df.columns if len(new_df[var].value_counts()) > 15]
target_variables=[var for var in new_df.columns if var not in excluded_variables]

# Defining r to autofit the number and size of plots
r = int(len(target_variables)/3 +1)
````

````python
# Defining a function to Notate the percent count of each value on the bars
def annot_percent(axes):
    '''Takes axes as input and labels the percent count of each bar in a countplot'''
    for p in plot.patches:
        total = sum(p.get_height() for p in plot.patches)/100
        percent = round((p.get_height()/total),2)
        x = p.get_x() + p.get_width()/2
        y = p.get_height()
        plot.annotate(f'{percent}%', (x, y), ha='center', va='bottom')
````

````python
# Plotting the countplots for each variable in target_variables
plt.figure(figsize=(18,r*3))
for n,var in enumerate(target_variables):
    plot = plt.subplot(r,3,n+1)
    sns.countplot(x=new_df[var]).margins(y=0.15)
    plt.title(f'{var.title()}',weight='bold')
    plt.tight_layout()
    annot_percent(plot)
````
